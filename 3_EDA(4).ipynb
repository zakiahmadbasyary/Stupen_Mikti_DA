# FILTER

```python
df2= df[df["Continent"].str.contains("Oceania")]
df2.sort_values(by="World Population Percentage", ascending=False).head()
```

```python
df3=df[df["Rank"]<=10].sort_values(by="Rank", ascending=True)
df3
```

```python
df[df["Rank"] <=10].sort_values(by="Rank", ascending=True)
```

```python
filter_countries=['Nigeria','Pakistan','China']
df[df['Country'].isin(filter_countries)].sort_values(by="World Population Percentage", ascending=True)
```

```python
filter_rank= [i for i in range(1, 101) if i % 3 == 0]
df[df['Rank'].isin(filter_rank)].sort_values(by="World Population Percentage", ascending=False)
```

```python
df[df['Country'].str.contains('za')]
```

```python
df[df['Country'].str.startswith('I') & df['Country'].str.endswith('a')]
```

# GROUP BY

```python
df4=df.groupby("Continent")[['2022 Population','2020 Population']].sum().sort_values(by="2022 Population", ascending=True)
df4
```

```python
df5=df4.transpose()
df5
```

```python
df6=df.groupby("Continent")[['1970 Population','1980 Population','1990 Population','2000 Population','2015 Population','2020 Population','2022 Population']].sum().sort_values(by="Continent", ascending=True)
df6
```

```python
df7=df6.transpose()
df7.plot()
```
