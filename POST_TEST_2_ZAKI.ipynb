## DATA CLEANSING DAN PREPOCESSING

Mengubah kolom yang memuat data percentage ke float

```python
# Mengecek kolom yang objek
df.select_dtypes(include='object').info()
```

```python
# Mengubah tipe data dari float dan menghilangkan % serta mengganti nilai "-" dengan 0.00

#Kolom T_winrate
df['T_WinRate'] =  df['T_WinRate'].str.replace('-', '0.00')
df['T_WinRate'] = df['T_WinRate'].str.replace('%', '').astype(float)

#Kolom T_pickpercentage
df['T_PickPercentage'] =  df['T_PickPercentage'].str.replace('-', '0.00')
df['T_PickPercentage'] = df['T_PickPercentage'].str.replace('%', '').astype(float)

#Kolom Bs_winrate
df['BS_WinRate'] =  df['BS_WinRate'].str.replace('-', '0.00')
df['BS_WinRate'] = df['BS_WinRate'].str.replace('%', '').astype(float)

#Kolom Rs_winrate
df['RS_WinRate'] =  df['RS_WinRate'].str.replace('-', '0.00')
df['RS_WinRate'] = df['RS_WinRate'].str.replace('%', '').astype(float)

#Kolom Ban_percentage
df['T_BansPercentage'] =  df['T_BansPercentage'].str.replace('-', '0.00')
df['T_BansPercentage'] = df['T_BansPercentage'].str.replace('%', '').astype(float)

#Kolom Pick&Ban percentage
df['T_PicksBansPercentage'] =  df['T_PicksBansPercentage'].str.replace('-', '0.00')
df['T_PicksBansPercentage'] = df['T_PicksBansPercentage'].str.replace('%', '').astype(float)
```

# Exploratory Data Analysis

### Distribution of Win Rates

```python
df["T_WinRate"].describe()
```

```python
# Membuat histogram untuk melihat distribusi data numerical yaitu winrate

plt.figure(figsize=(10, 6))  # Atur ukuran grafik
sns.histplot(df['T_WinRate'], bins=20, kde=True,color="orange")  # bins: jumlah bar, kde: kurva distribusi
plt.title('Distribution of Win Rates', fontsize=16) #judul
plt.xlabel('Win Rate (%)', fontsize=12) # label koordinat x
plt.ylabel('Frequency', fontsize=12)  #label koordinat y
plt.show()
```

```python
# Membuat Violin plot untuk melihat statistika dan kepadatan datanya

plt.figure(figsize=(10, 6)) # ukuran figure nya

# Mencari statistikanya
mean_winrate = df['T_WinRate'].mean()
min_winrate = df['T_WinRate'].min()
max_winrate = df['T_WinRate'].max()

#Membuat violin plot
sns.violinplot(y=df['T_WinRate'], color="orange")
plt.title('Distribution of Win Rates', fontsize=16)
plt.ylabel('Win Rate (%)', fontsize=12)
plt.ylim(-10, 110)

# Menampilkan data label statistika di kanan atas
plt.scatter(x=[0], y=mean_winrate, color='blue', label=f'Mean: {mean_winrate:.2f}', zorder=5)
plt.scatter(x=[0], y=min_winrate, color='green', label=f'Min: {min_winrate}', zorder=5)
plt.scatter(x=[0], y=max_winrate, color='red', label=f'Max: {max_winrate}', zorder=5)
plt.legend(title='Statistics', loc='upper right')
plt.show()
```

### Win Rate distribution each hero

```python
# sorting winrate secara ascanding
df_winrate=df.sort_values(by="T_WinRate", ascending=True)
df_winrate[["Hero", "T_WinRate"]]
```

### Win rate Blue side (bs) & Red sise(rs)

```python
# Membuat dataframe baru untuk memuat kolom hero, Bs_winrate, dan Rs_winrate
winrate_data=df[["Hero","BS_WinRate","RS_WinRate"]]
winrate_data = winrate_data.set_index("Hero")
winrate_data
```

```python
# Membuat barplot dengan 3 kolom data menggunakan pyplot express

# melakukan melt (membuat data dari bentuknya lebar ke bentuk yang panjang agar mudah dibuat barplot)
df_melted = pd.melt( winrate_data.reset_index(), id_vars='Hero',  value_vars=["BS_WinRate","RS_WinRate"],  var_name='Winrate Type',  value_name='Win Rate'  )

# Membuat barplotnya
fig = px.bar(df_melted, x="Hero", y="Win Rate", color='Winrate Type', barmode='group',title='Win rate Blue side (bs) & Red sise(rs)')
# Mengatur layoutnya
fig.update_layout(yaxis_title='Hero',xaxis_title='Win rate',xaxis_tickangle=90,width=2000,height=800)
fig.show()
```

### Distribution of hero roles

```python
# Membuat dataframe baru yang memuat kolom roles dengan jumlah hero yang termasuk rolesnya

df_roles= df.groupby('Roles')['Hero'].count().reset_index().sort_values(by='Hero', ascending=False)
df_roles
```

```python
# Membuat barchart dengan seaborn

plt.figure(figsize=(18, 6))
sns.barplot(x="Roles", y='Hero', data=df_roles, palette='viridis')
plt.title('Distribution of Hero Roles', fontdict={'fontsize': 18}, color='black',fontweight='bold')
plt.xlabel('Hero Roles', fontdict={'fontsize': 14}, color='black')
plt.ylabel('Count', fontdict={'fontsize': 14}, color='black')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Membuat Piechart untuk membandingkan percentagenya menggunakan seaborn
plt.figure(figsize=(8, 8))
colors = ['#B4A0D1', '#8DA9D3', '#70B5B5', '#60C76D', '#B4D65E', '#F6EB61','gray']

#Membagi data yang nilai hero kecil dibuat others
small_roles = df_roles[df_roles['Hero'] < 8]
large_roles = df_roles[df_roles['Hero'] >= 8]

# Menggabungkan
others = pd.DataFrame([{'Roles': 'Others', 'Hero': small_roles['Hero'].sum()}])
large_roles = pd.concat([large_roles, others])

plt.pie(large_roles["Hero"], labels=large_roles['Roles'], autopct='%1.1f%%', startangle=90,colors=colors)
plt.title('Distribution of Hero Roles')  # Tambahkan judul
plt.show()  # Tampilkan chart
```

### Correlation between hero picks & win rates


```python
# Membuat dataframe yang memuat data T_winrate dan T_pickpercentage
corr_piks_winrate=df[["T_WinRate","T_PickPercentage"]]
corr_piks_winrate
```

```python
# Membuat perhitungan korelasi antar kolom T_winrate dan T_pickpercentage
df_corr=corr_piks_winrate.corr()
df_corr
```

```python
#Membuat heatmap untuk menampilkan korelasi dalam bentuk grafiks

plt.figure(figsize=(12, 10))  # Atur ukuran figure
sns.heatmap(df_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation between hero picks & win rates')  # Tambahkan judul
plt.show()  # Tampilkan heatmap
```

```python
# Membuat scatter plot untuk melihat persebaran dan korelasinya
plt.figure(figsize=(10, 6))  # Atur ukuran figure
sns.scatterplot(x='T_PickPercentage', y='T_WinRate', data=corr_piks_winrate)
plt.title('Scatter Plot of Win Rate vs. Pick Percentage',fontdict={'fontsize': 15},fontweight='bold')  # Tambahkan judul
plt.xlabel('Pick Percentage (%)')  # Label sumbu x
plt.ylabel('Win Rate (%)')  # Label sumbu y
plt.show()  # Tampilkan scatter plot
```
