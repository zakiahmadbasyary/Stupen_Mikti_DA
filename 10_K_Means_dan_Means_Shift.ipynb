# MACHINE LEARNING

## K-MEANS CLUSTERING

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Buat data sintetis dengan 5 cluster

X, y_true = make_blobs(n_samples=5000, centers=5, cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], c=y_true)
plt.title('Data Asli dengan Label Cluster')
plt.show()

# Inisialisasi model K-Means dengan 5 cluster
kmeans = KMeans(n_clusters=5)

# Melatih model K-Means pada data X
kmeans. fit(X)

# Prediksi label cluster untuk setiap sampel data
y_pred = kmeans.predict(X)


from sklearn.metrics import accuracy_score
# Hitung akurasi model

accuracy = accuracy_score(y_true, y_pred)
print ('Akurasi:', accuracy)

plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title('Hasil Clustering dengan K-Means')
plt.show()
```

## METODE ELBOW UNTUK MENENTUKAN JUMLAH CLUSTERING

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Inisialisasi data
X, _ = make_blobs(n_samples=5800, centers=5, cluster_std=8.68, random_state=0)

inertia = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=8)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia)
plt.title('Metode Elbow')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Inersia')
plt.show()
```

## Mean Shift Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift
from sklearn.datasets import make_blobs
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import silhouette_score

centers = [[1, 1], [-1, -1], [1, -1]]
X, _ = make_blobs(n_samples=5000, centers=centers, cluster_std=0.6)
ms = MeanShift(bandwidth=1.2)
ms.fit(X)
labels = ms.labels_
score = silhouette_score(X, labels)

print("silhouette score:", score)

plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("Mean shift Clustering")
plt.show()

```

```python
# percobaan 2 denagn bandwith berbeda
ms2 = MeanShift(bandwidth=1)
ms2.fit(X)
labels2 = ms2.labels_
score2 = silhouette_score(X, labels2)

print("silhouette score:", score2)

plt.scatter(X[:,0], X[:,1], c=labels2)
plt.title("Mean shift Clustering")
plt.show()
```
