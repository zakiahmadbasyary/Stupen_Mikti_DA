## POPULASI DAN SAMPEL


```python
import numpy as np

# Sintesa populasi
population = np.random.normal(loc = 170, scale = 10, size = 10000)

# Sintesa sampel
sample = np.random.choice(population, size=100, replace = False)

# Rata- rata populasi dan sampel
population_mean = np.mean(population)
sample_mean = np.mean(sample)

# Rata-rata populasi dan sampel
print("Rata-rata populasi:", population_mean)
print("Rata-rata sampel:", sample_mean)
```

## Teorema Limit Pusat dan Hukum Bilangan Besar

Teorema Limit Pusat

```python
import numpy as np
import matplotlib.pyplot as plt

# Buat populasi sintetis dengan distribusi eksponsial
population = np.random.exponential(scale = 100, size = 10000)

# Ambil banyak sampel dari populasi
sample_means= []
num_samples= 1000
sample_size = 50
for _ in range(num_samples):
    sample = np.random.choice(population, size = sample_size, replace = False)
    sample_means.append(np.mean(sample))

# plot distribusi rata-rata sampel
plt.hist (sample_means,  bins=20, density=True)
plt.xlabel('Rata-rata Sampel')
plt.ylabel('Frekuensi')
plt.title('Demonstrasi Teorema limit Pusat')
plt.show()
```

Hukum Bilangan Besar



```python
import numpy as np
import matplotlib.pyplot as plt

# Buat populaso sintesis dengan distribusi normal
population = np.random.normal(loc =100, scale=15 , size=10000)
population_mean= np.mean(population)

# Htung rata-rata sampel dengan berbagai ukuran sampel
sample_sizes = np.arange(1,1001)
sample_means = []

for size in sample_sizes:
    sample = np.random.choice(population, size=size, replace=False)
    sample_means.append(np.mean(sample))

# Plot distribusi rata-rata sampel
plt.plot(sample_sizes, sample_means, label ='Rata-rata sampel')
plt.axhline(y=population_mean, color='r', linestyle='--', label='Rata-rata populasi')
plt.legend()
plt.xlabel('Ukuran Sampel')
plt.ylabel('Rata-rata Sampel')
plt.title('Demonstrasi Hukum Bilangan Besar')
plt.show()
```

## Estimasi Parameter: Titik dan Interval

```python
import numpy as np
import matplotlib.pyplot as plt

# Buat populasi sintesis dengan distrbusi normal
population = np.random.normal(loc =100, scale=15, size=10000)

# Estimasi titik
sample_mean = np.mean(population)

# Estimasi interval dengan tingkat keyakinan 95%
confidence_interval = np.percentile(population, [2.5, 97.5])

#  Plot distribusi pipulasi dan interval estimasi
plt.hist(population, density=True)
plt.axvline(x=sample_mean, color='r', linestyle='--', label='Estimasi Titik')
plt.axvspan(xmin=confidence_interval[0], xmax=confidence_interval[1] , color='g', alpha=0.5, label='Estimasi Interval')
plt.legend()
plt.xlabel('Tinggi Badan')
plt.ylabel('Densitas')
plt.title('Estimasi Titik dan Interval')
plt.show()
```

## Uji Hipotesis: Uji Z, Uji T, Uji Chi-Square, dan Uji F


Uji Hipotesis t

```python
import numpy as np
from scipy.stats import ttest_ind

# Data tinggi badan mahasiswa Universitas A
height_a = np.array([170,175,180,172,178])

# Data tinggi badan mahasiswa Universitas B
height_b = np.array([168,172,174,170,176])

# Uji hipotesis t
t_stat, p_value= ttest_ind(height_a, height_b)

# Cetak hasil uji hipotesis
print("Nilai t-statistik:", t_stat)
print("Nilai p-value:", p_value)

# Interpretasi hasil
if p_value < 0.05:
    print("Hipotesis nol ditolak. Ada perbedaan rata-rata tinggi badan antara Universitas A dan B.")
else:
    print("Hipotesis nol diterima. Tidak ada perbedaan rata-rata tinggi badan antara Universitas A dan B.")
```
